```python
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mean
```

```python
filepath = '/workspace/results_final/inception_lstm_no_threshold/training.csv'
df = pd.read_csv(filepath)
```

```python
# For training.csv
epochs = df['epoch'].tolist()
train_loss = df['loss'].tolist()
train_acc = df['accuracy'].tolist()
val_loss = df['val_loss'].tolist()
val_acc = df['val_accuracy'].tolist()
```

```python
plt.plot(epochs, train_acc , 'b', label='Training Accuracy') 
plt.plot(epochs, val_acc, 'y', label='Validation Accuracy') 
plt.title('Inception LSTM Inject Model without threshold \n Training Accuracy vs. Validation Accuracy')
plt.legend(loc="lower right")
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```

```python
plt.plot(epochs, train_loss , 'b', label='Training Loss') 
plt.plot(epochs, val_loss, 'y', label='Validation Loss') 
plt.title('Inception LSTM Inject Model without threshold \n Training Loss vs. Validation Loss')
plt.legend(loc="upper right")
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.show()
```
